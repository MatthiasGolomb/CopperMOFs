```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from matplotlib.backends.backend_pdf import PdfPages
from scipy import constants
import matplotlib.image as mpimg
%pylab inline
```

```python
C_up_proj = np.genfromtxt('Species_DOS/C_l_proj_dos_spin_up_raw.dat', unpack=True)
H_up_proj = np.genfromtxt('Species_DOS/H_l_proj_dos_spin_up_raw.dat', unpack=True)
N_up_proj = np.genfromtxt('Species_DOS/N_l_proj_dos_spin_up_raw.dat', unpack=True)
O_up_proj = np.genfromtxt('Species_DOS/O_l_proj_dos_spin_up_raw.dat', unpack=True)
Cu_up_proj = np.genfromtxt('Species_DOS/Cu_l_proj_dos_spin_up_raw.dat', unpack=True)

C_down_proj = np.genfromtxt('Species_DOS/C_l_proj_dos_spin_down_raw.dat', unpack=True)
H_down_proj = np.genfromtxt('Species_DOS/H_l_proj_dos_spin_down_raw.dat', unpack=True)
N_down_proj = np.genfromtxt('Species_DOS/N_l_proj_dos_spin_down_raw.dat', unpack=True)
O_down_proj = np.genfromtxt('Species_DOS/O_l_proj_dos_spin_down_raw.dat', unpack=True)
Cu_down_proj = np.genfromtxt('Species_DOS/Cu_l_proj_dos_spin_down_raw.dat', unpack=True)

Img335 = mpimg.imread('Cubefiles_Gamma/cube_017_eigenstate_00335_spin_1_k_point_0001.png')
Img340 = mpimg.imread('Cubefiles_Gamma/cube_027_eigenstate_00340_spin_1_k_point_0001.png')
Img341 = mpimg.imread('Cubefiles_Gamma/cube_029_eigenstate_00341_spin_1_k_point_0001.png')
#Img342 = mpimg.imread('Cubefiles/cube_011_eigenstate_density_00342_spin_1_k_point_0001.png')

Bands1 = np.genfromtxt('Bands118/band_structure.dat', unpack=True)
Bands2 = np.genfromtxt('Bands661/band_structure.dat', unpack=True)
```

```python
gs = gridspec.GridSpec(2,6)
fig = plt.figure(figsize=(20,10))

ax1 = plt.subplot(gs[0,0:2])
ax1.plot(C_up_proj[0], C_up_proj[2], label='s_up')
ax1.plot(C_up_proj[0], C_up_proj[3], label='p_up')
ax1.plot(C_up_proj[0], C_up_proj[4], label='d_up')
ax1.plot(C_down_proj[0], C_down_proj[2], label='s_down')
ax1.plot(C_down_proj[0], C_down_proj[3], label='p_down')
ax1.plot(C_down_proj[0], C_down_proj[4], label='d_down')
ax1.legend()
ax1.set_xlim(-8,-2)
ax1.set_ylim(0,10)
ax1.set_xlabel("eV")
ax1.set_title("Projected DOS of C in MOF-649")

ax2 = plt.subplot(gs[0,2:4])
ax2.plot(H_up_proj[0], H_up_proj[2], label='s_up')
ax2.plot(H_up_proj[0], H_up_proj[3], label='p_up')
ax2.plot(H_down_proj[0], H_down_proj[2], label='s_down')
ax2.plot(H_down_proj[0], H_down_proj[3], label='p_down')
ax2.legend()
ax2.set_xlim(-8,-2)
ax2.set_ylim(0,10)
ax2.set_xlabel("eV")
ax2.set_title("Projected DOS of H in MOF-649")

ax3 = plt.subplot(gs[0,4:6])
ax3.plot(N_up_proj[0], N_up_proj[2], label='s_up')
ax3.plot(N_up_proj[0], N_up_proj[3], label='p_up')
ax3.plot(N_up_proj[0], N_up_proj[4], label='d_up')
ax3.plot(N_down_proj[0], N_down_proj[2], label='s_down')
ax3.plot(N_down_proj[0], N_down_proj[3], label='p_down')
ax3.plot(N_down_proj[0], N_down_proj[4], label='d_down')
ax3.legend()
ax3.set_xlim(-8,-2)
ax3.set_ylim(0,10)
ax3.set_xlabel("eV")
ax3.set_title("Projected DOS of N in MOF-649")

ax4 = plt.subplot(gs[1,1:3])
ax4.plot(O_up_proj[0], O_up_proj[2], label='s_up')
ax4.plot(O_up_proj[0], O_up_proj[3], label='p_up')
ax4.plot(O_up_proj[0], O_up_proj[4], label='d_up')
ax4.plot(O_down_proj[0], O_down_proj[2], label='s_down')
ax4.plot(O_down_proj[0], O_down_proj[3], label='p_down')
ax4.plot(O_down_proj[0], O_down_proj[4], label='d_down')
ax4.legend()
ax4.set_xlim(-8,-2)
ax4.set_ylim(0,10)
ax4.set_xlabel("eV")
ax4.set_title("Projected DOS of O in MOF-649")

ax5 = plt.subplot(gs[1,3:5])
ax5.plot(Cu_up_proj[0], Cu_up_proj[2], label='s_up')
ax5.plot(Cu_up_proj[0], Cu_up_proj[3], label='p_up')
ax5.plot(Cu_up_proj[0], Cu_up_proj[4], label='d_up')
ax5.plot(Cu_down_proj[0], Cu_down_proj[2], label='s_down')
ax5.plot(Cu_down_proj[0], Cu_down_proj[3], label='p_down')
ax5.plot(Cu_down_proj[0], Cu_down_proj[4], label='d_down')
ax5.legend()
ax5.set_xlim(-8,-2)
ax5.set_ylim(0,10)
ax5.set_xlabel("eV")
ax5.set_title("Projected DOS of Cu in MOF-649")

plt.savefig('Species_DOS.png', dpi=300)
```
